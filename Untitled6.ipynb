```python
import theano
from theano import *
import theano.tensor as T
import numpy as np
import pandas as pd
from theano import function
```

```python
v1=T.scalar()
v2=T.scalar()
print(v1)
print(v2)
```

```python
s=v1-v2
```

```python
a=v1+v2
```

```python
cs=theano.function([v1,v2],s)
ca=theano.function([v1,v2],a)
```

```python
cs(50,23)
```

```python
ca(13,12)
```

```python
x=T.dmatrix('x')
y=T.dmatrix('y')
```

```python
z=x+y
```

```python
func=function([x,y],z)
```

```python
m1=[
    [1,2],
    [3,4]
]
```

```python
m2=[
    [6,8],
    [9,2]
]
```

```python
func(m1,m2)
```

```python
a=T.dmatrix('a')
sigmoid=1/(1+T.exp(-a))
af=function([a],sigmoid)
m3=[[1,2],[3,4]]
af(m3)
```

```python
from matplotlib import pyplot as plt
plt.scatter(m3,af(m3))
```

```python
from matplotlib import pyplot as plt
plt.scatter(m1,af(m1))
```
