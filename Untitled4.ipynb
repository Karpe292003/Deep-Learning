```python
import tensorflow as tf
import numpy as np
```

```python
a=tf.constant(15)
b=tf.constant(20)
```

```python
x=np.random.rand(10)
print(x)
```

```python
z=np.random.rand(5,4)
print(z)
```

```python
y=x*0.2 + 0.2
```

```python
print(y)
```

```python
w=tf.Variable(tf.random.normal([1]))
b=tf.Variable(tf.zeros([1]))
```

```python
print(w)
```

```python
print(b)
```

```python
def mse_loss():
  yp=w*x+b
  loss=tf.reduce_mean(tf.square(yp-y))
  return loss
```

```python
optimizer=tf.keras.optimizers.Adam()
```

```python
for step in range(15000):
  optimizer.minimize(mse_loss,var_list=[w,b])
  if step % 500 == 0:
    print(step,w.numpy(),b.numpy())
```
