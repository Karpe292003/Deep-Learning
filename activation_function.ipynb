```python
z = np.arange(-15, 15.1, 0.25) 
```

```python
def sigmoid(z):
  val = 1 / (1+np.exp(-z))
  return val
```

```python
y = sigmoid(z)
```

```python
plt.plot(z, y)
```

```python
z = np.arange(-5, 5.1, 0.25) 
```

```python
def relu(z):
  if z > 0:
    return z
  else:
    return 0
```

```python
y = [relu(n) for n in z]
```

```python
plt.plot(z, y)
```

```python
def softmax(z):
  val = np.exp(z) / sum(np.exp(z))
  return val
```

```python
data = [45,56,12,36,73,22]
```

```python
softmax(data)
```

```python
data[np.argmax(softmax(data))]
```

```python
def htan(z):
  val = (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
  return val
```

```python
y = htan(z)
```

```python
plt.plot(z, y)
```

```python
plt.plot(z, sigmoid(z), color = 'r', label='sigmoid')
plt.plot(z, htan(z), color='g', label='tanh')
plt.grid()
plt.legend()
```
